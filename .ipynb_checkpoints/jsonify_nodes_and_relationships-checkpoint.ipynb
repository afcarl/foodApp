```python
from flask import Flask, request, render_template, send_from_directory,jsonify
import code
from py2neo import authenticate, Graph
import py2neo
import os
import glob
import json
# import jsonify
from os import listdir
from py2neo import Graph, Node, NodeSelector
from neo4j.v1 import GraphDatabase
```

```python
driver = GraphDatabase.driver("bolt://localhost:7687", auth=("neo4j", "food"))
session = driver.session()

node_result = list(session.run("MATCH (a) WITH collect({node_id: id(a),node_type:labels(a)[0],node_name:a.name}) AS nodes RETURN nodes"))
rel_result = list(session.run("MATCH (a)-[r]->(b) where b.name = 'garlic' WITH collect({source_id: id(a), target_id: id(b),rel_id: id(r),rel_properties:properties(r),rel_type:type(r)}) AS edges  RETURN edges"))

node_list = node_result[0].values()
rel_result = rel_result[0].values()
```
